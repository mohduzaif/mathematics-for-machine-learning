### QQ Plot. 

```python
import numpy as np  
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = sns.load_dataset('iris')
```

```python
data.head()
```

```python
sns.kdeplot(data = data['sepal_length'])
```

```python
sepal_length = data['sepal_length'].to_list()
```

```python
x_percentile = [] 
for i in range(1, 101): 
    x_percentile.append(np.percentile(np.sort(sepal_length), i)) 
```

```python
normal_sample = np.random.normal(loc=10, scale=1, size=1000)
```

```python
normal_sample.mean(), normal_sample.std()
```

```python
y_percentile = [] 
for i in range(1, 101): 
    y_percentile.append(np.percentile(np.sort(normal_sample), i))
```

```python
sns.scatterplot(x=x_percentile, y=y_percentile)
```

#### `NOTE` We have one built-in(statemodel) library for QQ plot. 

```python
# using statsmodel. 
from statsmodels.graphics.gofplots import qqplot
from scipy import stats
```

```python
# create a QQ plot between two set of data. 
qqplot(data['sepal_length'], dist=stats.norm, line='45')
```

```python
import statsmodels.api as sm

# Example data
X = np.random.rand(100)
y = 3 * X + np.random.normal(0, 1, 100)

# Fit regression
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

# QQ plot of residuals
sm.qqplot(model.resid, line='45')
plt.title("QQ Plot of Regression Residuals")
plt.show()
```

### Pareto Distribution. 

```python
# apply the formula for pareto distribution. 

x = np.linspace(0.1, 10, 100)

# create a parameter in pareto distribution. 
xm = 1 
alpha = 3 

# Calculate the y values of the Pareto distribution
y = alpha * (xm**alpha) / x**(alpha + 1) 

plt.plot(x, y) 
plt.show()
```

```python
log_x = np.log(x) 
log_y = np.log(y) 

plt.plot(log_x, log_y)
```

```python
# the line is straight line so that distribution is pareto distribution. 
```

`NOTE : ` We can also prrof it using QQ plot, by just comparing both the distribution. 

```python
data = np.sort(np.linspace(1, 10, 1000))
normal_data = np.sort(np.random.normal(loc=10, scale=1, size=1000))

x_percentile = [] 
y_percentile = []

for i in range(1, 101): 
    x_percentile.append(np.percentile(np.sort(data), i))
    y_percentile.append(np.percentile(np.sort(normal_data), i)) 

plt.scatter(x_percentile, y_percentile)
```
