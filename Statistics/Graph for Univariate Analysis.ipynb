```python
import numpy as np  
import pandas as pd 

import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
titanic_df = pd.read_csv('./Datasets/Titanic.csv') 
```

```python
titanic_df.head()
```

### Barplot  for Categorical Columns.

```python
sns.countplot(x='Sex', data=titanic_df, palette='coolwarm')
```

```python
# using matplotlib 
count_col = titanic_df['Sex'].value_counts()

plt.bar(count_col.index, count_col.values, edgecolor = 'black') 
plt.xlabel('Sex')
plt.ylabel('Count of each sex')
plt.title('Count of Male and Female')

plt.show()
```

### Pie Chart for Categorical columns. 

```python
sex_counts = titanic_df['Sex'].value_counts() 
sex_counts
```

```python
plt.pie(
    sex_counts, 
    labels=sex_counts.index, 
    autopct='%1.1f%%'
)

plt.title("Passenger Distribution by Sex (Seaborn Styled)")
plt.show()
```

```python
# cummulative frequency. 
sns.histplot(
    data=titanic_df,
    x='Age',
    cumulative=True,    # cumulative frequency
    stat='density',     # or 'count' if you want actual cumulative counts
    bins=30,
    kde=True,
    color='skyblue'
)

plt.title("Cumulative Frequency of Age (Seaborn)")
plt.xlabel("Age") 
plt.ylabel("Cumulative Density")
plt.show()
```
