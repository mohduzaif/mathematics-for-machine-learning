```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt 
import seaborn as sns 
```

#### Here we are generating a data using uniform distribution.# here we are generating a data using uniform distribution.

```python

num_of_sample = 10000
sample_size = 300 
distribution_range = (0, 1) 

samples = np.random.uniform(distribution_range[0], distribution_range[1], (num_of_sample, sample_size))
```

```python
samples.shape
```

```python
samples
```

```python
sample_means = np.mean(samples, axis=1)
```

```python
plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.title('Histogram of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
sns.kdeplot(sample_means)
```

#### Here we are generating a data using using Explontial Distribution.

```python
# expontial parameters. 
sample_size = 100 
num_of_sample = 10000
lambda_para = 2 

expontial_samples = np.random.exponential(scale=1/lambda_para, size=(num_of_sample, sample_size))
```

```python
expontial_samples.shape
```

```python
expontial_samples
```

```python
sample_means = np.mean(expontial_samples, axis=1)
```

```python
sample_means.shape
```

```python
plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.title('Histogram of Sample Mean')
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.show()
```

```python
sns.kdeplot(x=sample_means)
```

### Here we are doing same task for multiple distribution. 

```python
# Set the parameters 
num_of_sample = 10000 
sample_size = 100 

# Poisson distribution parameters 
labmda_rate = 2 

# Gamma distribution parameters 
alpha = 2
theta = 2

# Binomial distribution parameters 
num_trails = 10 
success_probability = 0.6 

# Generate samples from the distributions. 
poisson_samples = np.random.poisson(lam=lambda_para, size=(num_of_sample, sample_size))
gamma_samples = np.random.gamma(shape=alpha, scale=theta, size=(num_of_sample, sample_size))
binomial_samples = np.random.binomial(num_trails, success_probability, size=(num_of_sample, sample_size))

# Generate mean for each distribution. 
poisson_sample_means = np.mean(poisson_samples, axis=1)
gamma_sample_means = np.mean(gamma_samples, axis = 1)
binomial_sample_means = np.mean(binomial_samples, axis=1)
```

```python
# Plot the histograms of the sample means
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

axs[0].hist(poisson_sample_means, bins=30, density=True, edgecolor='black')
axs[0].set_title('Sample Distribution of sample mean')
axs[0].set_xlabel('Sample Mean')
axs[0].set_ylabel('Density')

axs[1].hist(gamma_sample_means, bins=30, density=True, edgecolor='black')
axs[1].set_xlabel('Sample Mean')
axs[1].set_ylabel('Density')

axs[2].hist(binomial_sample_means, bins=30, density=True, edgecolor='black')
axs[2].set_xlabel('Sample Mean')
axs[2].set_ylabel('Density')

plt.show()
```

## Small Case study of CLT on Titanic Dataset. 

```python
import pandas as pd 
import numpy as np  

import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
titanic_df = pd.read_csv('./Datasets/Titanic.csv')
```

```python
titanic_df.shape
```

```python
titanic_df = titanic_df.drop('Survived', axis=1)
```

```python
titanic_df
```

```python
titanic_df['Fare'].mean()
```

```python
fare_samples = [] 
for i in range(1, 101):
    samples = titanic_df['Fare'].sample(50).values.tolist()
    fare_samples.append(samples)
```

```python
len(fare_samples)
```

```python
samples = np.array(fare_samples)
```

```python
samples_mean = samples.mean(axis = 1)
```

```python
samples_mean
```

```python
samples_mean.shape
```

```python
plt.hist(samples_mean, bins=20, edgecolor = 'black', density=True)
plt.show()
```

```python
sns.kdeplot(x = samples_mean)
plt.show()
```

```python
print('Sampling distribution of sample mean', samples_mean.mean())
print('Sampling distribution of sample STD', samples_mean.std())
```

#### `NOTE : ` We are trying to says a interval to said about the sample mean, k ye value is range me hogi or wo hum calculate kar lenge using Empherical Rule(68-95-99 Rule). 

#### And this process of telling is called Confidence interval.

```python
lower_limit = samples_mean.mean() - 2 * samples_mean.std() / np.sqrt(50)
upper_limit = samples_mean.mean() + 2 * samples_mean.std() / np.sqrt(50)
```

```python
print('The interval in which the mean of sample is lie ', lower_limit, '-', upper_limit)
```

```python
print('The mean and standard deviation of the population is ', titanic_df['Fare'].mean(), 'and', np.sqrt(titanic_df['Fare'].std()))
```
