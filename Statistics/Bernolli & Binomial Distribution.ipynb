```python
import pandas as pd 
import numpy as np    

import matplotlib.pyplot as plt 
import seaborn as sns 
```

### Binomial Distribution. 

```python
# According to forluma we generate the data. 
number_of_trails = 10 
prob_of_success = 0.5
size = 1000 


binomial_dist = np.random.binomial(number_of_trails, prob_of_success, size)
```

```python
plt.hist(binomial_dist)
plt.show()
```

```python
sns.displot(x=binomial_dist)
plt.show()
```
